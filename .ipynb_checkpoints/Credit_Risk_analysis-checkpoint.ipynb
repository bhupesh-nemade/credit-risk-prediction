```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('credit_risk_500k.csv')
df
```

```python
df.dtypes
```

```python
#Check missing values
df.isnull().sum()
```

```python
#handle the missing value
df['Age'].fillna(df['Age'].median(),inplace=True)
df['Income'].fillna(df['Income'].median(),inplace=True)
df['Loan_Amount'].fillna(df['Loan_Amount'].median(),inplace=True)
```

```python
df.isnull().sum()
```

# Exploratory Data Analysis(EDA)

```python
sns.countplot(data=df,x='Loan_Status')
plt.title('Loan_Approval status count')
plt.xlabel('Loan appoval(1=Apporval,0=Not Approval)')
plt.ylabel('status')
plt.show()
```

```python
sns.countplot(data=df,x='Credit_History',hue='Loan_Status')
plt.title('Loan_Approval by credit history')
plt.xlabel('credit history (1=yes 0=No)')
plt.ylabel('status')
plt.show()
```

```python
sns.boxplot(x='Loan_Status',y='Income',data=df)
plt.title('Loan vs Income')
plt.xlabel('Loan')
plt.ylabel('Income')
plt.show()
```

```python
sns.countplot(data=df,x='Education',hue='Loan_Status')
plt.show()

sns.countplot(data=df,x='Employment_Status',hue='Loan_Status')
plt.show()

sns.countplot(data=df,x='Marital_Status',hue='Loan_Status')
plt.show()

```

```python
df_model=pd.get_dummies(data=df,columns=['Education','Marital_Status','Employment_Status'],drop_first=True)
df_model.columns
```

```python
x=df_model.drop(columns=['Customer_ID','Loan_Status'])
y=df['Loan_Status']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression(max_iter=1000)
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
acc=accuracy_score(y_test,y_pred)
cm=confusion_matrix(y_test,y_pred)
cr=classification_report(y_test,y_pred)
print(acc)
print(cm)
print(cr)
```

```python
sns.heatmap(cm,fmt='d',annot=True,cmap='Blues')
plt.title('confusion matrix')
plt.xlabel('predicted')
plt.ylabel('Actual')
plt.show()
```

# Step 9: Create a Risk Score using Model Probabilities


```python
y_prob=model.predict_proba(x_test)[:,1]  # 1 means only include 1 column ie income

result=x_test.copy()
result['Actual_Loan']=y_test
result['Predicted_Loan']=y_pred
result['Loan_Risk_Score']=1-y_prob
```

```python
result.to_csv('Credit_Risk_Analysis_Result.csv',index=False)
```
